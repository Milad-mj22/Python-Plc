```python
from time import sleep

from opcua import Server

import random

```

```python
server=Server()

server.set_endpoint("opc.tcp://127.0.0.1:8081")

server.register_namespace('Room1')
```

```python
objects=server.get_objects_node()

objects
```

```python
tempsensors = objects.add_object('ns = 3 ; s="TS1"',"Tempreture Sensor 1")

tempsensors

```

```python

tempsensors.add_variable('ns = 3 ; s="TS1_VendorName"',"TS1_VendorName","Sensor King")
```

```python
tempsensors.add_variable('ns = 3 ; s="TS1_SerialNumber"',"TS1_SerialNumber",123456)
```

```python
temp=tempsensors.add_variable('ns=2;s="TS1_Tempreture"',"TS1_Tempreture",20)
```

```python
bulb = objects.add_object(2,"light bulb")
```

```python
bulb
```

```python
temp
```

```python
state = bulb.add_variable(2,"state bulb",False)
```

```python
state
```

```python
state.set_writable()
```

```python
tempreture=20

try:
    print('start server')
    server.start()
    print('server online')

    while True:
        tempreture+= 1

        temp.set_value(tempreture)

        print('new Temp'+str(temp.get_value()))

        print('state bulb'+str(state.get_value()))

        sleep(2)

finally:

    server.stop()

    print('server offline')


```
